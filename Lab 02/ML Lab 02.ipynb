```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def _details(self):
        print('Name:', self.name)
        print('Age:',self.age)
        
p1 = Person("John", 36)
p1._details()
```

```python
class Student(Person):
    def __init__(self, name, age, university):
        super().__init__(name, age)
        self.university = university
    
    def details(self):
        super()._details()
        print('University:',self.university)
        
student = Student("Shehryar Sohail", 21, "Fast-nu")
student.details()
```

```python
tups = [(2, 4), (3, 1), (5, 7)]

for i in range(len(tups)):
    for j in range(0, len(tups)-i-1):
        if tups[j][-1] > tups[j+1][-1]:
            tups[j], tups[j+1] = tups[j+1], tups[j]
            
tups
```

```python
elements = [1, 1, 1, 2, 3, 4, 5, 6, 6, 6, 7, 7, 8, 8, 9, 9, 0]
newList = []
for i, num in enumerate(elements):
    if i == len(elements)-1:
        break
    if not num == elements[i+1]:
        newList.append(num)
        
if elements[-1] != elements[-2]:
    newList.append(elements[-1])
        
newList
```
